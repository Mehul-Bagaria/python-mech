## Projectile Motion
The Python code for finding the trajectory, range, time of flight and maximum height of the projectile under the influence of gravitational field.  
**Equations :**  

$ \text { Time of Flight : } t=\frac{2 v_0 \sin \theta}{g} $  
$\text { Maximum height reached }: H= h + \frac{v_0^2 \sin ^2 \theta}{2 g}$  
$\text { Horizontal Range }: R= v_0\cos(\theta)  t $   
$\text { Trajectory formula}: y= h + x \tan \theta-\frac{g x^2}{2 v^2 \cos ^2 \theta}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def projectile_motion():
    try:
        # Take input values
        u = input("Enter the initial velocity : ")
        theta = input("Enter the angle of launch in degree : ")
        h = input("Enter the initial height : ")
        print("------------------\n")

        # Converting str into int
        u = int(u)
        theta = int(theta)
        h = int(h)

        # Defining value of g
        g = 9.8 #m/s^2


        # Converting the value of theta from degree to radians
        theta = theta* np.pi / 180

        # Time of flight to cover the symmetric part of the parabolic trajectory
        time_flight = 2*u*np.sin(theta) / g

        # Time of fight to cover the additional height
        t_add = (1/g) * ( -u*np.sin(theta) + np.sqrt( (u*np.sin(theta))**2 + 2*g*h ))

        # Define the maximum height and horizontal range
        max_height = h + (u * np.sin(theta))**2 / (2*g)
        two_theta = 2*theta
        horiz_range = u*np.cos(theta) * (time_flight + t_add)
        max_height_x = u*np.cos(theta)*(time_flight/2)
        max_height_y = max_height

        # Print the results
        print("\nThe horizontal range is: ", horiz_range)
        print("The maximum vertical height is :", max_height)
        print("The total time of flight is :", time_flight + t_add)


        # Evaluate the coordinates of the trajectory
        t = 0
        T = []
        X = []
        Y = []

        for i in range(101):
            x = u*np.cos(theta)*t
            y = h + x* np.tan(theta) - g*x**2/ (2*(np.cos(theta)*u)**2)
            dt = (time_flight + t_add)/100
            t = t + dt
            T.append(t)
            X.append(x)
            Y.append(y)


        # Plot the results
        plt.plot((horiz_range), (0), 'o', color = 'r', label = 'Range')
        plt.plot((max_height_x), (max_height_y), 'o', color = 'g', label = 'Max Height')
        plt.title('Projectile Motion')    
        plt.plot(X,Y, label = 'Trajectory')
        plt.legend()
        
    except Exception as e:
        print(e)
        print("Sorry, that's not a valid input !")
```

```python
projectile_motion()
```
